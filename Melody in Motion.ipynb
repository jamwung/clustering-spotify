```python
# Load the Files and Merge
files = glob.glob('./data/*.csv')

# merge all datafiles, append a decade marker
df = pd.DataFrame()
for file in files:
    df_loop = pd.read_csv(file)
    df_loop['decade'] = file.split('/')[-1][11:14]
    df = pd.concat([df, df_loop])

# drop duplicates, keep first instance
df.drop_duplicates(subset='uri', keep='first', inplace=True)
```

```python
# display health
info_dict = {'feature': df.columns, 'data_type': df.dtypes,
             'non-null_count': df.count()}

df_info = pd.DataFrame(info_dict).reset_index(drop=True)

toc.add_table(df_info,
              'Information Table of Features with Pandas Datatype'
              ' and Valid Data',
              index=False,
              border=False,
              preview=False)
```

<div style="text-align: justify;
            font-size:15px;">
    <b>Insights:</b>
    <br><br>
    Checking the over-all health/density of the dataset. There no null values.
</div>

```python
# set decade marker as a categorical variable, for sorting
decades = ['60s', '70s', '80s', '90s', '00s', '10s']
df['decade'] = pd.Categorical(df['decade'], categories=decades)
df.sort_values('decade', inplace=True)
```

```python
# plotting number of tracks & number of unique artists per decade
tracks_per_decade = df.groupby('decade')['uri'].count()
unique_artists_per_decade = df.groupby('decade')['artist'].nunique()

fig, axs = plt.subplots(1, 2, figsize=(15, 8))

tracks_per_decade.plot(kind='bar', color='#1DB954', ax=axs[0])
axs[0].set_title('Number of Tracks per Decade', fontsize=16)
axs[0].set_xlabel('Decade')
axs[0].set_ylabel('Count')
axs[0].tick_params(axis='x', rotation=0)
axs[0].spines['top'].set_visible(False)
axs[0].spines['right'].set_visible(False)

unique_artists_per_decade.plot(kind='bar', color='#48494B', ax=axs[1])
axs[1].set_title('Number of Unique Artists per Decade', fontsize=16)
axs[1].set_xlabel('Decade')
axs[1].set_ylabel('Count')
axs[1].tick_params(axis='x', rotation=0)
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)

toc.add_fig('Plotting Trends of Number of Track'
            'and Unique Artists per Decade',
            width=100)
```

<div style="text-align: justify;

    <br><br>
    Observing some temporal trends in the dataset. It can be observed that the number of tracks experienced a dip during the 90's. However, the Number of Unique Artists are in upward trend, increasing as time passes.
</div>

```python
# spotify artists in the dataset
artist_names = ' '.join(df['artist'].str.lower().dropna().astype(str))

wordcloud = WordCloud(width=800, height=400,
                      background_color='white').generate(artist_names)

plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')

plt.axis('off') 
plt.title('Spotify Artists', fontsize=18)

toc.add_fig('Word Cloud Visualization of Spotify Artists', width=100)
```

        Statistiscal Testing and Transformations
    </h3>
</div>

***
<div style="text-align: justify;
            font-size:15px;">
Since the team aims to perform clustering using GMM, categorical features were not utilized and continuous features were prepared. In this section, the team automated the Statistical Testing process and applied Log Transformation if necessary. 
</div>

```python
# to automate normal tests + transformations
transformations = {
    'log': np.log,
    'reciprocal': lambda x: 1 / x,
    'boxcox': stats.boxcox
}

df_transformed = df.copy()
transformed_features = {}

for col in cont_columns:
    data = df[col].copy()
        
    _, p_value = stats.normaltest(df[col])
    
    if p_value < 0.05:
        for transformation, transform_func in transformations.items():
            try:
                if transformation == 'log':
                    transformed_ft = transform_func(data + 1e-10)
                else:
                    transformed_ft = transform_func(data)
                df_transformed[col + '_' + transformation] = transformed_ft
                
                transformed_features[col] = {
                    'transformation': transformation,
                    'transformed_feature': transformed_ft
                }
                
                break  
            except (ValueError, ZeroDivisionError):
                continue 
    else:
        transformed_features[col] = {
            'transformation': 'none',
            'transformed_feature': df[col]
        }

df_transformation = (
    pd.DataFrame()
    .from_dict(transformed_features, orient='index')
    .drop(columns='transformed_feature')
    .rename_axis('feature')
)
# for col, info in transformed_features.items():
#     print(f"Feature: {col}, Transformation: {info['transformation']}")
toc.add_table(df_transformation,
              'Summary of Transformation Done based from Statistical Test',
              index=True,
              border=False,
              preview=False)
```

<div style='text-align: justify; font-size:15px;'>

    <br><br>
    All the continuous features were Log Transformed.

By this point, the data is now ready for clustering. The team employed the use of <span style="background-color:#dddddd">GridSearchCV</span> to optimize the implementation of <span style="background-color:#dddddd">GaussianMixture</span> with respect to the clusters' <span style="background-color:#dddddd">silhouette_score</span>. The team included <span style="background-color:#dddddd">Standard Scaler</span> in the pipeline as standard scaling ensures that all features have similar scales and prevents variables with larger magnitudes from dominating the clustering process. Pickle files were loaded for streamlined run of analysis.

```python
# Plotting the Data
pca = PCA(random_state=1337)
df_pca = pca.fit_transform(df_transformed[trans_columns])
fig, ax = plt.subplots(figsize=(15,8))
sns.scatterplot(
    x=df_pca[:, 0], y=df_pca[:, 1],
    color='#1DB954', ec='#191414'
)
plt.xlabel('PC1')
plt.ylabel('PC2')

toc.add_fig('Scatter Plot of Transformed Data using'
            ' First and Second Principal Components',
            width=100)
```

</div>

```python
# Performing Grid-Search
def silhouette_scorer(estimator, X):
    """
    Scoring function for silhouette_score inside GridSearchCV.
    """
    labels = estimator.predict(X)
    score = silhouette_score(X, labels)
    return score

rng = np.random.RandomState(143)

param_grid = dict(
    clusterer__n_components=range(5, 11),
    clusterer__reg_covar=[1e-6, 1e-5, 1e-4],
    clusterer__init_params=['kmeans', 'random_from_data'],
)

fixed_params = dict(
    covariance_type='full',
    max_iter=100,
    n_init=1,
    random_state=rng,
    verbose=0,
    verbose_interval=20,
)

scoring = {'silhouette': silhouette_scorer}

gmm = GaussianMixture(**fixed_params)

scaler = StandardScaler()
pipe = Pipeline([('scaler', scaler), ('clusterer', gmm)])

gs = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=[(slice(None), slice(None))],
    scoring=scoring,
    refit='silhouette',
    verbose=3,
)
```
